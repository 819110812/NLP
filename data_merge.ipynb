```python
import pandas as pd 
documents = pd.read_json("dataset\document.json",orient="records",lines=True,encoding="utf-8")
documents
```

```python
documents['id'] = documents["id"].apply(lambda x: x.strip())
```

```python
del documents["_id"]
```

```python
documents
```

```python
abstract = pd.read_json("dataset/abstract.json",orient="records",lines=True,encoding="utf-8")
```

```python
import re
pattern = re.compile(r'PMC\d*')
```

```python
abstract['id'] = abstract['id'].apply(lambda x: pattern.findall(x)[0])
```

```python
del abstract['_id']
```

```python
abstract
```

```python
documents.set_index("id",inplace=True)
documents
```

```python
abstract.set_index("id",inplace=True)
abstract
```

```python
documents = documents.join(abstract,lsuffix="_doc",rsuffix="_abst")
```

```python
documents["abstract"]
```

```python
#coding: UTF-8
str(documents["abstract"][0])
```

```python
def clean_text(content):
    if "AB  -" in content:
        content = content.replace("AB  -","")
    return content

documents["abstract"] = documents["abstract"].apply(clean_text)
```

```python
#coding=utf-8
documents["abstract"][0]
```

```python
documents.to_json("documents.json",orient="records",lines=True)
```
